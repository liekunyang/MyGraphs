```python
def rolling_linear_reg( x, y,window=6):
        r = np.zeros(len(x))
        r[:] = np.NAN
        myrange = range(int(window/ 2), len(x) - int(window / 2))

        a = np.vstack([x, np.ones(len(x))]).T
        for i in myrange:
            r[i] = np.linalg.lstsq(a[i:i + window], y[i:i + window])[0][0]
        return r
```

```python
def derive( df):
    for m in [44, 45, 47, 49]:
        df.loc[:, 'd' + str(m) + 'dt'] = rolling_linear_reg(df['time'], df['Mass' + str(m)])

    return df
```
